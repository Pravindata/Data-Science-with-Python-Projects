```python
df=df.drop(['Over18','EmployeeNumber','EmployeeCount','StandardHours'],axis=1)
```

```python
df.columns
```

```python
df.shape
```

```python
df[['RelationshipSatisfaction','JobSatisfaction','EnvironmentSatisfaction','JobInvolvement']]
```

```python
df["TotalSatisfaction mean"]=(df['RelationshipSatisfaction']+df['JobSatisfaction']+df['EnvironmentSatisfaction']+df['JobInvolvement']+df['WorkLifeBalance'])/5
```

```python
df.head()
```

```python
def Satif(df):
    if df['TotalSatisfaction mean'] > 2.35:
        return 1
    else:
        return 0
    
df['Satif']=df.apply(lambda df:Satif(df),axis=1)

df['Satif']
```

```python
df.shape
```

```python
df["JobSatisf mean"]=(df['JobSatisfaction']+df['JobInvolvement'])/2
```

```python
df.shape
```

```python
def MovingPeople(df):
    if df["NumCompaniesWorked"] > 4:
        return 1
    else:
        return 0
    
    
df["MovingPeople"] = df.apply(lambda df:MovingPeople(df), axis=1)
df["MovingPeople"]
```

```python
df.shape
```

```python
def LongDis(df):
    if df['DistanceFromHome'] > 11:
        return 1
    else:
        return 0
    
    
    
df["LongDis"]=df.apply(lambda df:LongDis(df),axis = 1)
df["LongDis"]
        
```

```python
def MiddleTraining(df):
    if df["TrainingTimesLastYear"] >=3 and df["TrainingTimesLastYear"] <= 6:
        return 1
    else:
        return 0
    
    
df['MiddleTraining']=df.apply(lambda df:MiddleTraining(df), axis=1)
df["MiddleTraining"]
```

```python
df["Time_in_each_comp"]=(df['Age']-20)/(df['NumCompaniesWorked']+1)
df["Time_in_each_comp"]
```

```python
df.shape
```

```python
numeric_df=df.select_dtypes(include=[np.number])
categoric_df=df.select_dtypes(exclude=[np.number])
```

```python
print("Numeric Type : ", numeric_df.columns)
print("\n")
print("Categoric Type : ",categoric_df.columns)
```
