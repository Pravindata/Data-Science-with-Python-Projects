```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randomnumber=np.random.rand(10)
```

```python
randomnumber
```

```python
style.use('ggplot') #plot style

plt.plot(randomnumber,'g',label='Line one',linewidth=2) 
#

plt.xlabel("Range") # X axis name or label
plt.ylabel("Number") #Y axis name or label
plt.title("First plot") #Title of graph

plt.legend() #Legend label or name of line 
plt.show()  # To display the graph with all datas
```
