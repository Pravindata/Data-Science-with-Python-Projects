```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
causes="Chronic Diseases","unitentional injuries","Alzheimers","Infuenza","Sepsis","Other"
percentile=[62,5,4,2,1,26]
```

```python
plt.figure(figsize=(10,10))
explode=(0.03,0,0,0,0,0)
#plt.pie(percentile,labels=causes,explode=explode,autopct="%1.1f%%",startangle=70)
plt.pie(percentile,labels=causes,explode=explode,autopct="%1.1f%%",startangle=70)
plt.axis("equal")
plt.title("Ohio state -2012: Leading causes of death ")
plt.show()
```
